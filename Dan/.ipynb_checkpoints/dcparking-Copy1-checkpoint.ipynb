```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
#connect to local db
engine = create_engine("mysql://root:Password@127.0.0.1/DCDATA_DB")
rds_connection_string = "root:Password@127.0.0.1/DCDATA_db?charset=utf8" 
```

```python
#check table names
engine.table_names()
```

```python
#load parking ticket csv
csv_file = "Resources/Parking_Violations_DC_2018.csv"
dcparking_df = pd.read_csv(csv_file)
dcparking_df.head()
```

```python
#use pandas to load csv to database
dcparking_df.to_sql(name='PARKING', con=engine, if_exists='replace', index=False)
```

```python
#confirm data loaded successfully
pd.read_sql_query('select * from PARKING', con=engine).head(5)
```

```python
#create new columns
newdcpark_df = dcparking_df[['X', 'Y', 'VIOLATION_DESCRIPTION', 'TICKET_ISSUE_DATE']]
newdcpark_df.head()
```

```python
# clean up date to make it just month and day
dates = []
dates = newdcpark_df['TICKET_ISSUE_DATE'].values
dates [1]
```

```python
# get the month

MM = []
for d in dates:
    MM.append(d[5:7])
print(MM)
```

```python
#get the day
DD = []
for d in dates:
    DD.append(d[8:10])
print(DD)
```

```python
#create the MONTH column and populate (you'll get an error but it works)
newdcpark_df["MONTH"]=MM
```

```python
#create the DAY column and populate (you'll get an error but it works)
newdcpark_df["DAY"]=DD
```

```python
#create new columns (getting rid of the ticket_issue_date)
finaldcpark_df = newdcpark_df[['X', 'Y', 'VIOLATION_DESCRIPTION', 'MONTH', 'DAY']]
finaldcpark_df.head()
```

```python
#reload new data after dropping columns from Parking
newdcpark_df.to_sql(name='PARKING', con=engine, if_exists='replace', index=False)
```
